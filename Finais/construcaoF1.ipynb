```python
import pandas as pd
import numpy as np

# Read CSV files
cenario9 = pd.read_csv("C:/Users/Utilizador/Desktop/Pedro/Universidade/4º Ano/1º Semestre/AASE/P2/Cenario9Sem_IDS.csv")
predicts = pd.read_csv("C:/Users/Utilizador/Desktop/Pedro/Universidade/4º Ano/1º Semestre/AASE/P2/PredictsTrain.csv")
```

```python
# Combine dataframes
final = pd.concat([cenario9, predicts['prediction']], axis=1)

# Initialize new columns
final['precisaoNew'] = 0
final['prejuizoNew'] = 0
```

```python
# Calculate variacaoPrecisao
final['variacaoPrecisao'] = abs((final['resale_price_Lakh'] - final['prediction']) / final['resale_price_Lakh']) * 100

# Calculate precisaoOld and prejuizoOld
final['precisaoOld'] = (final['variacaoPrecisao'] > 8).astype(int)
final['prejuizoOld'] = final['precisaoOld'] * final['resale_price_Lakh'] / final['prediction']
```

```python
# Normalize variacaoPrecisao
variacao_precisao_normalizada = (final['variacaoPrecisao'] - final['variacaoPrecisao'].min()) / (final['variacaoPrecisao'].max() - final['variacaoPrecisao'].min())

# Assign values to satisfacaoOld based on precisaoOld condition
fator_escala = 25
valores_brutos_satisfacao_old = np.where(final['precisaoOld'] == 1, 30 + variacao_precisao_normalizada * 30 * fator_escala, 61 + variacao_precisao_normalizada * 39 * fator_escala)
final['satisfacaoOld'] = np.where(final['precisaoOld'] == 1, np.minimum(valores_brutos_satisfacao_old, 50), np.minimum(valores_brutos_satisfacao_old, 100))

# Calculate values for satisfacaoNew
valores_brutos_satisfacao_new = np.where(final['precisaoOld'] == 1, final['satisfacaoOld'] + (final['satisfacaoOld'] * 0.3 + variacao_precisao_normalizada * 0.1) * fator_escala, final['satisfacaoOld'])
final['satisfacaoNew'] = np.minimum(valores_brutos_satisfacao_new, 100)
```
