```python
import pandas as pd

# Carregar os DataFrames a partir dos arquivos CSV
df = pd.read_csv('test.csv')
dft = pd.read_csv('todascenario2.csv')

# Criar um dicionÃ¡rio de mapeamento de city para id_city em dft
mapeamento_id_cidade = dict(zip(dft['city'], dft['id_city']))

# Substituir os valores de city por seus IDs correspondentes em df
df['ID'] = df['city'].map(mapeamento_id_cidade)

# Visualizar o DataFrame df atualizado
print(df)
```
